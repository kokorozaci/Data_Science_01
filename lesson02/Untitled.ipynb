### Задание 2

```python
x = np.linspace(0, 10, 51)
f = np.cos(x)
plt.plot(x, f, color="green")
plt.title("График f(t)")
plt.xlabel("Значения t")
plt.ylabel("Значения f")

plt.axis([0.5, 9.5, -2.5, 2.5])

plt.grid(color="lightgrey")
plt.show()
```

### Задание 3

```python
x = np.linspace(-3, 3, 51)
y1 = x*2
y2 = 2*x+0.5
y3 = -3*x-1.5
y4=np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_title("График y1")
ax1.set_xlim([-5, 5])
ax2.plot(x, y2)
ax1.set_title("График y2")
ax3.plot(x, y3)
ax1.set_title("График y3")
ax4.plot(x, y4)
ax1.set_title("График y4")
```

### Задание 4

```python
import pandas as pd
```

```python
creditcard = pd.read_csv('creditcard.csv')
creditcard.head()
```

```python
class_counts = creditcard['Class'].value_counts()
```

```python
plt.style.use('fivethirtyeight')
class_counts.plot(kind="bar")
plt.show()
```

```python
class_counts.plot(kind="bar", logy=True)
plt.show()
```

```python
plt.hist(creditcard.loc[creditcard["Class"] == 0]['V1'], bins=20, alpha=0.5, color='red', density=True)
plt.hist(creditcard.loc[creditcard["Class"] == 1]['V1'], bins=20, alpha=0.5, color='gray',  density=True)
plt.xlabel("Class")
plt.legend(labels=["Class 0", "Class 1"])
plt.show()
```
